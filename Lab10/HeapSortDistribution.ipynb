# Experimental analysis of the Heap Sort execution time

```
def max_heapify(a, index, size):
    left = 2 * index + 1
    right = 2 * index + 2
    largest = index

    if left < size and a[left] > a[largest]:
        largest = left
    if right < size and a[right] > a[largest]:
        largest = right

    if largest != index:
        a[index], a[largest] = a[largest], a[index]
        max_heapify(a, largest, size)

def build_max_heap(a):
    n = len(a)
    for i in range(n // 2 - 1, -1, -1):
        max_heapify(a, i, n)

def heap_sort(a):
    build_max_heap(a)
    n = len(a)
    for i in range(n - 1, 0, -1):
        a[0], a[i] = a[i], a[0]
        max_heapify(a, 0, i)
    return a
```

```
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 21, 1)

plt.plot(n, 5 * n - 4,  label='Heap sort best time $5n-4$')
plt.plot(n, 3 / 4 * n ** 2 + 17 / 4 * n - 4, label='Heap average time $3/4 n^2+17/4 n - 4$')
plt.plot(n, 3 / 2 * n ** 2 + 7 / 2 * n - 4, label='Heap worst time $3/2 n^2+7/2 n - 4$')
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc='upper left')
plt.show()
```

```
def max_heap_steps(a, index, steps):
    izq, der = 2 * index, (2 * index) + 1
    steps += 1
    if izq < len(a) and a[izq] > a[index]:
        max_idx = izq
        steps += 1
    else:
        max_idx = index
        steps += 1
    if der < len(a) and a[der] > a[max_idx]:
        max_idx = der
        steps += 1
    if max_idx != index:
        a[index], a[max_idx] = a[max_idx], a[index]
        steps += 2
        steps = max_heap_steps(a, max_idx, steps)
    return steps

def build_max_heap_steps(a, steps):
    for i in range(int(len(a) / 2), -1, -1):
        steps += max_heap_steps(a, i, steps)
    return steps

def heap_sort_steps(a):
    size = len(a)
    steps = 1
    sorted_array = []
    steps += build_max_heap_steps(a, steps)
    while size > 1:
        maxim = a[1]
        a[1] = a[size - 1]
        # del a[-1]
        size -= 1
        sorted_array.insert(0, maxim)
        steps += max_heap_steps(a, 1, steps)
        steps += 6
    if len(a) == 1:
        sorted_array.append(a[0])
        steps += 2
    return steps
```

Generate all permutations without repetitions of size *n* (for small *n*), plot the "exact" time  distribution (number of steps), and print the media and the standar deviation

```
import itertools
import matplotlib.pyplot as plt
import numpy as np

def generate_permutations(n):
    elements = list(range(1, n+1))
    permutations = list(itertools.permutations(elements))
    return permutations

def plot_time_distribution(permutations):
    num_steps = [i for i in range(1, len(permutations)+1)]
    plt.plot(num_steps, 'o-')
    plt.xlabel('Permutation Index')
    plt.ylabel('Number of Steps')
    plt.title('Exact Time Distribution')
    plt.show()

def calculate_mean_std(permutations):
    num_steps = len(permutations)
    mean = np.mean(num_steps)
    std_dev = np.std(num_steps)
    return mean, std_dev

# Generate permutations
n = 4
permutations = generate_permutations(n)

# Plot time distribution
plot_time_distribution(permutations)

# Calculate mean and standard deviation
mean, std_dev = calculate_mean_std(permutations)
print("Mean:", mean)
print("Standard Deviation:", std_dev)
```

# Sample of runs (montecarlo)
Generate a sample of *runs=1000,10000* random permutations of size *n=3,4,30,100* (for large *n*), plot the "approximated o estimated" time  distribution (number of steps), and print the media and the standar deviation.

```
import numpy as np
import time
import matplotlib.pyplot as plt

def calculate_time_distribution(n, runs):
    time_distributions = []
    for _ in range(runs):
        start_time = time.time()
        np.random.permutation(n)
        end_time = time.time()
        time_distributions.append(end_time - start_time)

    return time_distributions

# Set the parameters
n_values = [3, 4, 30, 100]
runs = [1000, 10000]

# Generate time distributions for different values of n and runs
for n in n_values:
    for run in runs:
        time_distributions = calculate_time_distribution(n, run)

        # Plot the time distribution
        plt.hist(time_distributions, bins=30)
        plt.xlabel('Time (seconds)')
        plt.ylabel('Frequency')
        plt.title(f'Time Distribution (n={n}, runs={run})')
        plt.show()

        # Calculate and print mean and standard deviation
        mean = np.mean(time_distributions)
        std_dev = np.std(time_distributions)
        print(f"Mean (n={n}, runs={run}): {mean:.6f}")
        print(f"Standard Deviation (n={n}, runs={run}): {std_dev:.6f}")
        print('-' * 40)
```
